```python
# Import modules
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

**Drought in 2010**

```python
# read data
dataframe = gpd.read_file('data/d2010j.shp')

# check the head row in dataframe
dataframe.head()
```

```python
# Create plot

#fig, ([ax1, ax4],[ax2,ax5],[ax3,ax6]) = plt.subplots(ncols=2, nrows=3, figsize=(15, 10), sharey=True)
fig, ([ax1,ax2, ax3],[ax4,ax5,ax6],[ax10,ax11,ax12]) = plt.subplots(ncols=3, nrows=3, figsize=(10, 10), sharey=True)

divider1 = make_axes_locatable(ax1)
divider2 = make_axes_locatable(ax2)
divider3 = make_axes_locatable(ax3)
divider4 = make_axes_locatable(ax4)
divider5 = make_axes_locatable(ax5)
divider6 = make_axes_locatable(ax6)

divider10 = make_axes_locatable(ax10)
divider11 = make_axes_locatable(ax11)
divider12 = make_axes_locatable(ax12)

cax1 = divider1.append_axes("right", size="5%", pad=0.1)
cax2 = divider2.append_axes("right", size="5%", pad=0.1)
cax3 = divider3.append_axes("right", size="5%", pad=0.1)
cax4 = divider4.append_axes("right", size="5%", pad=0.1)
cax5 = divider5.append_axes("right", size="5%", pad=0.1)
cax6 = divider6.append_axes("right", size="5%", pad=0.1)

cax10 = divider10.append_axes("right", size="5%", pad=0.1)
cax11 = divider11.append_axes("right", size="5%", pad=0.1)
cax12 = divider12.append_axes("right", size="5%", pad=0.1)

dataframe.plot(ax=ax1, column="3_Jun", cmap='RdYlBu', linewidth=.5, edgecolor="black" , vmin=-2, vmax=2, legend=True, cax=cax1, 
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax2, column="3_Aug", cmap='RdYlBu', linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax2, 
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax3, column="3_Nov", cmap='RdYlBu', linewidth=.5, edgecolor="black",   vmin=-2, vmax=2, legend=True, cax=cax3,
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax4, column="6_Jun", cmap='RdYlBu', linewidth=.5, edgecolor="black",  vmin=-2, vmax=2, legend=True, cax=cax4,
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax5, column="6_Aug", cmap='RdYlBu', linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax5, 
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax6, column="6_Nov", cmap='RdYlBu', linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax6,
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax10, column="12_Jun", cmap='RdYlBu',linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax10,
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax11, column="12_Aug", cmap='RdYlBu', linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax11, 
               legend_kwds={'label': "Drought Index"});

dataframe.plot(ax=ax12, column="12_Nov", cmap='RdYlBu', linewidth=.5, edgecolor="black", vmin=-2, vmax=2, legend=True, cax=cax12,
               legend_kwds={'label': "Drought Index"});


# Add titles
ax1.set_title("SRI3-Jun", fontsize=15)
ax2.set_title("SRI3-Aug", fontsize=15)
ax3.set_title("SRI3-Nov", fontsize=15)
ax4.set_title("SRI6-Jun", fontsize=15)
ax5.set_title("SRI6-Aug", fontsize=15)
ax6.set_title("SRI6-Nov", fontsize=15)

ax10.set_title("SRI12-Jun", fontsize=15)
ax11.set_title("SRI12-Aug", fontsize=15)
ax12.set_title("SRI12-Nov", fontsize=15)



# Remove axis
ax1.axis('off')
ax2.axis('off')
ax3.axis('off')
ax4.axis('off')
ax5.axis('off')
ax6.axis('off')

ax10.axis('off')
ax11.axis('off')
ax12.axis('off')

plt.tight_layout()
plt.savefig('d2010.jpeg')
plt.show();
```
